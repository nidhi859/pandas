```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame( {"a" : [1,2,3], "b" : [3,5,2]})
print(df)
print(df.dtypes)
print(df.describe())
print(df.min())
print(df.max())
print(df.corr())
print(df.count())
print(df.median())
```

```python
dict = {"Name":["ram","sham","atul"], "Marks":[20,19,13], "Age":[12,13,12]}

df = pd.DataFrame(dict)
print(df)
print(np.shape(df))
print(df.describe())
print(df["Name"])
print(df["Name"][0])
df["Name"][0] = "riya"
print(df)
df.index = [00, 11, 22]
print(df)
print(df["Marks"].dtypes)
```

```python
newdf = pd.DataFrame(np.random.rand(10,5), index = np.arange(10))
print(newdf)
print(newdf.dtypes)
# newdf[0][0] = "abc"
# print(newdf.dtypes)
print(newdf.index)
print(newdf.columns)
print(newdf.to_numpy())
print(newdf.T)
print(newdf.sort_index(axis=0, ascending = False)) #0 for rows
print(newdf.sort_index(axis=1, ascending = False))  #1 for columns
```

```python
newdf2 = newdf
print(type(newdf2)) # here newdf2 is a memoryview, that points to the newdf
newdf2[0][0] = 234
print(newdf2)
print(newdf)
newdf3 = newdf.copy() # here newdf3 is a copy of newdf
newdf3[0][0] =2345
print(newdf3)
print(newdf)
```
